### A3:

```python
import numpy as np
import time

size = 100000

big_python_list_without_Numpy = list(range(size))
big_python_array_with_Numpy = np.array(size)

start_time_big_python_list_without_Numpy = time.time()
for i in range(len(big_python_list_without_Numpy)):
    big_python_list_without_Numpy[i] = big_python_list_without_Numpy[i] **2
end_time_big_python_list_without_Numpy = time.time()

print("Time taken for big Python list without Numpy: ", end_time_big_python_list_without_Numpy - start_time_big_python_list_without_Numpy)

start_time_big_python_list_with_Numpy_end = time.time()
big_python_array_with_Numpy = big_python_array_with_Numpy **2
end_time_big_python_list_with_Numpy_end = time.time()

print("Time taken for big Python array with Numpy: ", end_time_big_python_list_with_Numpy_end - start_time_big_python_list_with_Numpy_end)
```

Ergebnis:
Numpy ist um einiges schneller, bei size = 1000 ca 0.0007 Unterschied (nicht bemerkbar), jedoch mit zunehmender Größe size = 100000 ca 0.02 Unterschied (mit noch größeren immer weiter auseinandergehend). Numpy hat kaum veränderungen bei der Laufzeit.
